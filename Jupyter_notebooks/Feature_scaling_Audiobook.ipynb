```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
outlier_free_data=pd.read_csv("../Datasets/audiobook_outlier_treated.csv")
```

```python
scaler=StandardScaler()
x=outlier_free_data.drop(columns=['Target'])
y=outlier_free_data['Target']
```

```python
#Applying standardinzation
x_scaled=scaler.fit_transform(x)

#Coverting back to dataframe
scaled_data=pd.DataFrame(x_scaled,columns=x.columns)
scaled_data['Target']=y
scaled_data.head()
```

```python
scaled_data.to_csv("audioboook_scaled.csv",index=False)
print("Feature scaling is complete and saved")
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

numeric_cols = outlier_free_data.drop(columns=['Target']).select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 4, i)
    sn.histplot(scaled_data[col], kde=True, bins=30, color='orange')
    plt.title(f"Distribution of {col} (Scaled)")
plt.tight_layout()
plt.show()
```
